## Project Scenario:

An international firm that is looking to expand its business in different countries across the world has recruited you. You have been hired as a junior Data Engineer and are tasked with creating a script that can extract the list of the top 10 largest economies of the world in descending order of their GDPs in Billion USD (rounded to 2 decimal places), as logged by the International Monetary Fund (IMF). 

The required data seems to be available on the URL mentioned below:

URL: https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29

```python
URL = "https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

```python
!pip install lxml
```

```python
import pandas as pd
import numpy as np
```

```python
# Extract tables from webpage using Pandas 
tables = pd.read_html(URL)
```

```python
type(tables)
```

```python
# We are interested in table 3
type(tables[3])
```

```python
df = tables[3]
```

```python
df
```

```python
df.ndim
```

```python
df.columns
```

```python
df.shape
```

```python
# Replace column headers with column numbers
df.columns = range(df.shape[1])
```

```python
df
```

```python
# Retain the columns with index 0 and 2 (name of country and value of GDP quoted by IMF)

df = df[[0,2]]
```

```python
df
```

```python
df.columns = ['Country','GDP (Million USD)']
```

```python
df
```

```python
df.info()
```

```python
df = df.iloc[1:11,:]
```

```python
df
```

```python
df
```

```python
# Change the data type of GDP (Million USD) column to integer

df['GDP (Million USD)'] = df['GDP (Million USD)'].astype(int)
```

```python
df
```

```python
# Top 10 GDPs of world in descending order
df = df.sort_values('GDP (Million USD)', ascending=False)
```

```python
df
```

```python
df.info()
```

```python
# Convert GDP value from Million to Billion USD

df[['GDP (Million USD)']] = df[['GDP (Million USD)']]/1000
```

```python
df
```

```python
# Round the values to 2 decimal places

df[['GDP (Million USD)']] = np.round(df[['GDP (Million USD)']],2)
```

```python
# Rename the column header

df.rename(columns = {'GDP (Million USD)':'GDP (Billion USD)'})
```

```python
# Load into csv
df.to_csv('./Largest_Economies.csv')
```
